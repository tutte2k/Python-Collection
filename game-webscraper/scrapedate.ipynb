```python
import pandas as pd
df1 = pd.DataFrame({
    'Name': ['Tutte', 'Tutte Druid', 'Tuttelito', 'Adversario'],
    'Online': [True, False, False, True],
    'Unix': ['1654034400', '1654034400', '1654034400', '1654034400']
})
df2 = pd.DataFrame({
    'Name': ['Tutte', 'Tutte Druid', 'Tuttelito', 'Adversario'],
    'Online': [False, True, False, True],
    'Unix': ['1654034460', '1654034460', '1654034460', '1654034460']
})
df3 = pd.DataFrame({
    'Name': ['Tutte', 'Tutte Druid', 'Tuttelito', 'Adversario'],
    'Online': [False, False, True, True],
    'Unix': ['1654034520', '1654034520', '1654034520', '1654034520']
})
df4 = pd.DataFrame({
    'Name': ['Tutte', 'Tutte Druid', 'Tuttelito', 'Adversario'],
    'Online': [False, False, False, True],
    'Unix': ['1654034580', '1654034580', '1654034580', '1654034580']
})
df = pd.concat([df1, df2, df3, df4], ignore_index=True)
new_df= df.assign(Timespan=df.groupby('Name').Unix.shift(-1))
new_df = new_df.dropna()
new_df['Timespan'] = new_df['Timespan'].astype(int)
new_df['Unix'] = new_df['Unix'].astype(int)
new_df['Duration'] = new_df['Timespan'] - new_df['Unix']
new_df
datetime_df = new_df
datetime_df['Unix'] =  pd.to_datetime(datetime_df['Unix'], unit='s')
datetime_df['Timespan'] =  pd.to_datetime(datetime_df['Timespan'], unit='s')
temp = {True:'blue', False:'red'}
datetime_df['Color'] = datetime_df['Online'].map(temp)
mock_datetime_df = datetime_df
fig = px.timeline(mock_datetime_df, x_start='Unix', x_end='Timespan', y='Name', color="Color")
fig.show()
```
