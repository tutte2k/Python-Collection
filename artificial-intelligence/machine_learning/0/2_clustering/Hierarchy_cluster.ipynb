```python
from sklearn.cluster import AgglomerativeClustering
import pandas as pd
import scipy.cluster.hierarchy as sch
```

```python
df = pd.read_csv('customers.csv')
X = df.iloc[:,2:].values
```

```python
agg= AgglomerativeClustering(n_clusters=None,distance_threshold=0.5,compute_full_tree=True,compute_distances=True)
agg.fit(X)
```

```python
agg.labels_
```

```python
agg.get_params()
```

```python
dgram = sch.dendrogram(sch.linkage(X,method='ward'))
```

```python
agg= AgglomerativeClustering(n_clusters=5)
aval = agg.fit_predict(X)
```
