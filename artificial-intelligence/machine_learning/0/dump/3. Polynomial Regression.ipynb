# Polynomial Regression

## Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

## Importing the dataset

```python
data = pd.read_csv('3. Position_Salaries.csv')
x = data.iloc[:, 1:-1].values
y = data.iloc[:, -1].values
```

## Training the Linear Regression model on the whole dataset

```python
reg = LinearRegression()
reg.fit(x,y)

y_pred = reg.predict(x)
```

## Training the Polynomial Regression model on the whole dataset

```python
from sklearn.preprocessing import PolynomialFeatures

poly_reg = PolynomialFeatures(degree=2)
x_poly = poly_reg.fit_transform(x)

reg2 = LinearRegression()
reg2.fit(x_poly,y)

y2_pred=reg2.predict(x_poly)
```

## Visualising the Linear Regression results

```python
plt.scatter(x,y ,c ='red')
plt.plot(x,y_pred)
plt.xlabel('position')
plt.ylabel('salary')
```

## Visualising the Polynomial Regression results

```python
plt.scatter(x,y ,c ='red')
plt.plot(x,y2_pred)
plt.xlabel('position')
plt.ylabel('salary')
```

## Visualising the Polynomial Regression results (for higher resolution and smoother curve)

```python
poly_reg3 = PolynomialFeatures(degree=13)
x3_poly = poly_reg3.fit_transform(x)
reg3 = LinearRegression()
reg3.fit(x3_poly, y)
y3_pred = reg3.predict(x3_poly)
plt.scatter(x, y, c='red')
plt.plot(x, y3_pred)
plt.xlabel('position')
plt.ylabel('salary')
```

## Predicting a new result with Linear Regression

```python
reg.predict([[6.5]])
```

## Predicting a new result with Polynomial Regression

```python
reg2.predict(poly_reg.fit_transform([[6.5]]))
```

```python
reg3.predict(poly_reg3.fit_transform([[6.5]]))
```
